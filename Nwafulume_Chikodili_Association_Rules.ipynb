```python
data = mydataset.pivot_table(index='TransactionNo', columns='Items', aggfunc={'Items':'count'}).fillna(0)
```

```python
#Generating association rules
Rules = apriori(data, min_support=0.02, use_colnames=True)
Rules1 = association_rules(Rules, metric='lift', min_threshold=1)
```
